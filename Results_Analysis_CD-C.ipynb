```python
import pandas as pd

url="https://raw.githubusercontent.com/pushpanjali316/test-repo/main/CD_Results.csv"
df=pd.read_csv(url)

print(df.head())
```

```python
df
```

```python
# Average marks for each subject
print("\nAverage marks per subject:")
print(df.mean(numeric_only=True))
print("Minimum Average : ",df.mean(numeric_only=True).min())
print("Maximum Average : ",df.mean(numeric_only=True).max())

# Highest marks in each subject
print("\nHighest marks per subject:")
print(df.max(numeric_only=True))

# Lowest marks in each subject
print("\nLowest marks per subject:")
print(df.min(numeric_only=True))
```

```python
"""Finding CGPA for each student"""
# Step 3: Credits for each subject
credits={
    'CD121': 3, 'CD122': 2, 'CD123': 2, 'CD124': 3,
    'CD125': 2, 'CD126': 4, 'CD161': 1, 'CD162': 1, 'CD163': 1,'CD164':1.5,'CD165':1
}

# Step 4: Function to get grade and grade point
def get_grade_point(marks):
    if marks >= 90:
        return "A+", 10
    elif marks >= 80:
        return "A", 9
    elif marks >= 70:
        return "B", 8
    elif marks >= 60:
        return "C", 7
    elif marks >= 50:
        return "D", 6
    elif marks >= 45:
        return "E", 5
    elif marks >= 40:
        return "P", 4
    else:
        return "F", 0

# Step 5: Calculate subject grades and CGPA
total_credits=sum(credits.values())
cgpa_list=[]

for idx, row in df.iterrows():
    total_points=0
    for subject,credit in credits.items():
        grade, gp=get_grade_point(row[subject])
        df.loc[idx, f"{subject}_Grade"]=grade
        total_points += gp * credit
    cgpa=total_points / total_credits
    cgpa_list.append(round(cgpa, 2))

df["CGPA"]=cgpa_list
df
```

```python
df1=df.sort_values(by='CGPA',ascending=False)
df1
```

📌 Note: How Ranking Works Here

Students with the same CGPA get the same rank.

We use method='dense' so there are no gaps in the ranking numbers.

Example: If three students are ranked 2, the next rank will be 3 (not 5).

Ranks are given in descending order of CGPA (highest CGPA = rank 1).



```python
df1['Rank']=df1['CGPA'].rank(method='dense', ascending=False)
df1[['Reg.No','CGPA','Rank']].head(10)
```
